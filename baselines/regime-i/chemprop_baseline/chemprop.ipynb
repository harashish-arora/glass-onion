```python
import pandas as pd
from sklearn.model_selection import train_test_split

seed = 101

# 1. Load Data
print("Loading ESOL dataset...")
esol_url = "https://raw.githubusercontent.com/deepchem/deepchem/master/datasets/delaney-processed.csv"
full_df = pd.read_csv(esol_url)

# 2. Rename columns to be consistent
full_df = full_df.rename(columns={
    "smiles": "SMILES",
    "measured log solubility in mols per litre": "LogS"
})

# 3. Perform the exact same split
train_df, test_df = train_test_split(full_df, test_size=0.2, random_state=seed)

# 4. Save to CSVs for Chemprop
train_df.to_csv('esol_train.csv', index=False)
test_df.to_csv('esol_test.csv', index=False)

print(f"Data saved: Train ({len(train_df)}), Test ({len(test_df)})")
```

```python
!chemprop train \
  --data-path esol_train.csv \
  --task-type regression \
  --smiles-columns SMILES \
  --target-columns LogS \
  --output-dir esol_checkpoints \
  --epochs 60
```

```python
!chemprop predict \
  --test-path esol_test.csv \
  --model-paths esol_checkpoints/model_0/best.pt \
  --o esol_predictions.csv \
  --smiles-columns SMILES
```

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score

# Load ground truth and predictions
truth = pd.read_csv('esol_test.csv')
preds = pd.read_csv('esol_predictions.csv')

y_true = truth['LogS']
y_pred = preds['LogS']

# Calculate metrics
mse = mean_squared_error(y_true, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_true, y_pred)

print("-" * 30)
print(f"CHEMPROP ESOL RESULTS")
print(f"RMSE: {rmse:.4f}")
print(f"R^2:  {r2:.4f}")
print("-" * 30)
```
